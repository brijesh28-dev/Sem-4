# Filtering

```python
data={
    "Name":["A","B","C","D","E","F"],
    "AGE":[22,24,26,27,28,30],
    "Gender":["Male","Female","Male","Male","Male","Female"],
    "Marks":[60,70,80,90,100,110],
    "Dept":["IT","HR","IT","Finance","HR","IT"],
}
df=pn.DataFrame(data)
df
```

```python
df[df['AGE']<25]
```

```python
df[(df['AGE']<25)&(df['Dept']=='IT')]
```

```python
df[(df['Marks']>=70)&(df['Dept']=='HR')]
```

```python
df[df['Gender']!="Female"]
```

```python
df[df['Dept'].isin(["IT","HR"])]
```

```python
df[df["AGE"].between(20,30)]
```

```python
df[(df['Marks']>=60)&(df['AGE'].between(20,30))]
```

# Quation on titanic

```python
import pandas as pn
import numpy as np
df=pn.read_csv("titanic.xls")
df
```

```python
# Q1 list of surviver
df[df["survived"]==1]
```

```python
# Q2 male pasanger
df[df["sex"]=="male"]
```

```python
# Q3 age > 30
df[df["age"]>30]
```

```python
# Q4 first calss pasanger 
df[df["pclass"]==1]
```

```python
# Q5 female surviver
df[(df["sex"]=="female")&(df["survived"]==1)]
```

```python
# Q6 age between 20 , 40
df[df["age"].between(20,40)]
```

```python
# Q7 male + age(20,40) + survived
df[(df["sex"]=="male")&(df["survived"]==1)&(df["age"].between(20,40))]
```

```python
# Q8 embark is in(5,6)
df[df['embarked'].isin(["S","C"])]
```
