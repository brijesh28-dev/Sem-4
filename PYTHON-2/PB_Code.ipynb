### 58 / Write Python code to remove outliers from any given DataFrame

```python
''' 59 / Consider the following data:
data = {
    "A": ["TeamA", "TeamB", "TeamB", "TeamC", "TeamA"],
    "B": [50, 40, 40, 30, 50],
    "C": [True, False, False, False, True]
}
Convert this to a Pandas DataFrame and remove duplicate rows from it. Reset index values.'''
import pandas as pn
import numpy as np
data = {
    "A": ["TeamA", "TeamB", "TeamB", "TeamC", "TeamA"],
    "B": [50, 40, 40, 30, 50],
    "C": [True, False, False, False, True]
}
df=pn.DataFrame(data)
df=df.drop_duplicates()
df.reset_index()
df
```
